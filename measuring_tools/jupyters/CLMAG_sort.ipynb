```python
import pandas as pd


blocks = 4
header = ["measure_day", "measure_month", "measure_year", "distribuidora", "comercialitzadora", "provincia", "agree_tensio", "agree_tarifa", "agree_dh", "agree_tipo", "magnitud"]
header += [str(x[1]) + str(x[0]) for x in enumerate(blocks * range(25))]
header.append('nan')

df_clmag = pd.read_csv('/home/puig/Fitxers de Medidas/ECASA Devel Medidas/OK/CLMAG_0034_201802_20180531.0', sep=';', names=header, dtype={'distribuidora': 'str', 'comercialitzadora': 'str', 'agree_tipo': 'str', 'measure_day': 'str', 'measure_month': 'str'})
df_clmag.drop('nan', axis=1, inplace=True)
df_clmag = df_clmag.sort_values(['comercialitzadora', 'agree_tensio', 'agree_tarifa', 'agree_dh', 'agree_tipo', 'measure_day', 'measure_month', 'measure_year'])
df_clmag.to_csv('/tmp/mel', sep=';', header=False, columns=header, index=False, line_terminator=';\n')
```
